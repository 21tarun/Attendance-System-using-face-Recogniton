```python
import face_recognition
import cv2
import numpy as np
```

```python
imgEmma = face_recognition.load_image_file('ImagesAttendance/Emma Watson.jpg')
imgEmma = cv2.cvtColor(imgEmma,cv2.COLOR_BGR2RGB)
imgTest = face_recognition.load_image_file('TestImages/emma_kristen2.png')
imgTest = cv2.cvtColor(imgTest,cv2.COLOR_BGR2RGB)
```

```python
faceLoc = face_recognition.face_locations(imgEmma)[0]

encodeEmma = face_recognition.face_encodings(imgEmma)[0]
cv2.rectangle(imgEmma,(faceLoc[3],faceLoc[0]),(faceLoc[1],faceLoc[2]),(255,0,255),2) # top, right, bottom, left
 
faceLocTest = face_recognition.face_locations(imgTest)[0]
encodeTest = face_recognition.face_encodings(imgTest)[0]
cv2.rectangle(imgTest,(faceLocTest[3],faceLocTest[0]),(faceLocTest[1],faceLocTest[2]),(255,0,255),2)
```

```python
results = face_recognition.compare_faces([encodeEmma], encodeTest)
faceDis = face_recognition.face_distance([encodeEmma], encodeTest)
print(results[0],faceDis)
cv2.putText(imgTest,f'{results} {round(faceDis[0],2)} ',(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,255),3)
```

```python
cv2.imshow('Emma',imgEmma)
cv2.imshow('Emma Test',imgTest)
cv2.imwrite('output3.jpg', imgTest)
cv2.waitKey(0)
```
